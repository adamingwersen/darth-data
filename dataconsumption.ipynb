```python
import matplotlib.pyplot as plt
fz = 24
plt.rcParams.update({'font.size': fz})
figx = (20,10)
```

```python
email = 75000 / 10**6 #7000 words in 75kb ;bytes
photo = 3*10**6 / 10**6 #3mb photo; 3M bytes
s5vid = 16*10**6 / 10**6 #5s video; 16M bytes
netflix = 2.2 * 1.5 * 10**9 / 10**6 #1.5hr HD movie
```

```python
D = {u'E-mail': email,
     u'HD iPhone Photo': photo,
     u'5sec HD iPhone Video':s5vid
    }
fig = plt.figure(figsize=figx)
plt.style.use('seaborn-deep')
plt.bar(range(len(D)), list(D.values()), align='center', alpha = 0.7)
plt.xticks(range(len(D)), list(D.keys()))
plt.xticks(rotation=10, fontsize = fz)
plt.ylabel("Size in MB")

fig.savefig('images/subset.png', bbox_inches='tight')
```

```python
D[u'1.5hr HD Netflix Movie'] = netflix
fig = plt.figure(figsize=figx)
plt.style.use('seaborn-deep')
plt.bar(range(len(D)), list(D.values()), align='center', alpha = 0.7)
plt.xticks(range(len(D)), list(D.keys()))
plt.xticks(rotation=10, fontsize = fz)
plt.ylabel("Size in MB")
fig.savefig('images/full.png', bbox_inches='tight')
```
